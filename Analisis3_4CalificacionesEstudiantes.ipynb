Q2. La linea media del diagrama en color verde, es la mediana o segundo cuartil. Este valor de aproximadamente 83 es el valor debajo del cual se ubica el 50% de la poblacion. En otras palabras, aproximadamente 1,000 estudiantes tienen un promedio no mayor de 83. En este caso la mediana es muy cercana al valor medio de 82.5.

Q3. La linea superior de la caja, es el tercer cuartil. Este valor de aproximadamente 89.7 es el valor debajo del cual se ubica el 75% de la poblacion o el valor por el cual se encuentra el 25% de la poblacion. En otras palabras, aproximadamente 500 estudiantes se encuentran con un promedio de entre 83 y 89.7 o practicamente 90. Y 500 estudiantes tendrian un promedio mayor a 90.

El rango intercuartil o IQR seria 89.7 - 75.57 = 14.13. En estadística descriptiva, el rango intercuartílico (IQR) es una medida de dispersión estadística, que es la extensión de los datos. 

Con el IQR el valor inferior o bigote inferior seria paproximadamente de 89.7- (1.5*14.13) = 68.5.  Este seria el valor minimo del primer cuartil. Los valores por debajo de este minimo son considerados como potenciales atipicos u outliers.

Similarmente el valor maximo del tercer cuartil o Q3 seria 89.7 + (1.5*14.13) = 110.89. Sin embargo el valor maximo segun el boxplot es de 100 y no se presentan outliers. Por lo que en cuanto a valores maximos de promedios pareceria no existir errores o valores mayores a 100.

<A name='Method1'> </A> <h3>METHOD 1 Standard Deviations to DETECTING OUTLIERS.</h3> <h4>It assumes normal distribution. </h4>

Si los datos estan normalmente distribuidos, entonces el 95% de los datos está dentro de 1.96 de la desviacion estándar de la media. Entonces podemos eliminar los valores por encima (TopRangeGrade = MeanGrade + StDGrade * 1.96) o por debajo de ese rango (LowRangeGrade = MeanGrade - StDGrade * 1.96).

```python
MeanGrade=df.grade.mean()
MeanGrade
```

```python
StDGrade = df.grade.std() #Calculates standar deviation of grade
StDGrade
```

```python
TopRangeGrade = MeanGrade + StDGrade * 1.96
TopRangeGrade
```

```python
LowRangeGrade = MeanGrade - StDGrade * 1.96
LowRangeGrade
```

<A name='Method2'> </A>  <h3>METHOD 2 Interquartile Range (IQR) to DETECTING OUTLIERS </h3>

##### Sobre el rango interquartil algo se habia avanzado en la seccion anterior de visualizacion de datos. Sin embargo ahora se vera con otro nivel de detalle.

```python
#Interquartile Range (IQR): El IQR es la diferencia entre 
#los valores de los quartiles del 25% y del 75% es decir: IQR=Q3-Q1.

#Cualquier valor por debajo de  Q1 - 1.5 x IQR
#o por arriba de  Q3 + 1.5 x IQR 
#son tratados como outliers y son eliminados.
```

#### Es posible que el segundo metodo sea mas recomendable ya que al parecer no requiere de una distribucion normal de los datos.

```python
Q1 = df.grade.quantile(.25)
Q1
```

```python
Q3 = df.grade.quantile(.75)
Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
LowRange = Q1 - IQR * 1.5
LowRange
```

```python
TopRange = Q3 + IQR * 1.5
TopRange
```

```python
FilteredDf2=df #copia df original para posteriormente filtrar datos
```

```python
FilteredDf2=FilteredDf2.drop(FilteredDf2[FilteredDf2.grade < LowRange].index)
FilteredDf2
```

#### Buscando ahora los registros por arriba del rango alto

```python
FilteredDf2[FilteredDf2.grade > TopRange]
```

Python for Data Science: The ultimate beginners' guide to learning python data science step by step. Independently Published.

<A name='Regression1'> </A> <h3>Regression1</h3>

En el modelado estadístico, el análisis de regresión es un proceso estadístico para estimar las relaciones entre variables. Esta es una forma elegante de decir que usamos la regresión para crear una ecuación que explique el valor de una variable dependiente en función de una o varias variables independientes (Henley y Wolf, 2018).

#### Para el modelado estadístico en este caso se utilizará la librería o módulo statsmodels abreviada como sm del cual se utilizará un modelo de regresion lineal simple por el método de mínimos cuadrados simples u ordinary least squares (OLS), los cuales de acuerdo con Sarkar, Bali,y Sharma,(2018) "se utilizan normalmente para obtener el mejor ajuste lineal durante el entrenamiento del modelo."

```python
import statsmodels.formula.api as sm
```

De esta librería de statsmodels se utiliza ols para crear una formula indicada por el usuario la cual será precisamente el modelo en este caso de regresión.

Debemos decidir en qué columnas queremos realizar la regresión y cuál es la variable dependiente (Henley y Wolf, 2018). Por ejemplo de querer intentar predecir la calificación en función de la edad, las horas de ejercicio y las horas de estudio:

El formato de fórmula muestra la variable dependiente a la izquierda de la tilde (~) y las variables independientes que queremos considerar a la derecha (Henley y Wolf, 2018).

```python
result = sm.ols(formula='grade ~ age + exercise + hours',data=df_grades).fit()
result.summary()
```

Si observa los resultados que obtiene del resumen, la R-cuadrada representa el porcentaje de la variación en los datos que la regresión puede explicar: 0.664, o 66.4 por ciento, es bueno, pero no estupendo (Henley y Wolf, 2018).

#### El valor p (representado aquí por el valor de P>|t|) representa la probabilidad de que la variable independiente NO tenga efecto sobre la variable dependiente (Henley y Wolf, 2018). En general mantener  valores de p por debajo del 5 por ciento (0.05) es una practica común para considerar las variables con influencia. Po lo que la única variable que NO indica tener efecto es la edad con un 59,5 por ciento.

La idea fundamental aquí es que la hipótesis usará la aproximación lineal, luego mapeará con una función logística para la predicción binaria(Swamynathan, 2019).


#### La ecuación de regresión lineal en este caso es 
### y = mx + c 

La regresión logística se puede explicar mejor en la razón de probabilidades. Las probabilidades de que ocurra un evento se definen como la probabilidad de que ocurra un evento dividida por la probabilidad de que ese evento no ocurra.

razón de probabilidades de pasar vs fallar = probabilidad (y = 1) / 1 − probabilidad (y = 1)

#### Un logit es la base logarítmica e(log) de las probabilidades, por lo que al usar un modelo logit:

### log(p / p(1 - p)) = mx + c

La siguiente figura muestra la ecuacion del modelo de regresion lineal y la ecuación del modelo de probabilidad de regresión logística (y = 1) = 1 / 1 + e−(mx + c)

<A name='PerformaceClass1'> </A> <h3>Evaluación del rendimiento de un modelo de clasificación</h3>

La matriz de confusión es la tabla que se utiliza para describir el rendimiento del modelo de clasificación. La siguiente figura muestra la matriz de confusión.

#### Verdadero=Positivo. Falso=Negativo
#### 1 (Predicho Falso, Actual Falso) Verdadero Negativo.  (TN):  FALSE real que se predijo como FALSE 
#### 2 (Predicho Verdadero, Actual Verdadero) Verdadero Positivo (TP): VERDADERO real que se predijo como VERDADERO
#### 3 (Predicho Verdadero, Actual Falso) Falso Positivo. (FP): FALSO real que se predijo como VERDADERO (error de tipo I)
#### 4 (Predicho Falso, Actual Verdadero) Falso Negativo (FN): VERDADERO real que se predijo como FALSO (error de tipo II)

### Idealmente, un buen modelo debería tener altos TN y TP y menos errores de tipo I y II.

La siguiente celda es el código de ejemplo para producir una matriz de confusión.

```python
from sklearn import metrics
# generate evaluation metrics
print ("Accuracy :", metrics.accuracy_score(y, modelLR2.predict(x)))
#print ("AUC :", metrics.roc_auc_score(y, modelLR2.predict_proba(x)[:,1]))
```

```python
# generate evaluation metrics
print ("Confusion matrix :",metrics.confusion_matrix(y, modelLR2.predict(x)))
#print ("classification report :", metrics.classification_report(y, modelLR2.predict(x)))
```

La siguiente Tabla describe las métricas clave derivadas de una matriz de confusión para medir el rendimiento del modelo de clasificación. 

Al tener la matriz de confusion y a partir de la imagen que indica cada posición, podemos manualmente obtener algunos valores que se indican en la tabla anterior. El hacer manualmente algo similar se sugiere en el libro de "Practical Statistics for Data Scientists" 

En el corazón de las métricas de clasificación se encuentra la matriz de confusión (Bruce, y otros, 2020). La matriz de confusión es una tabla que muestra el número de predicciones correctas e incorrectas clasificadas por tipo de respuesta (ídem). Hay varios paquetes disponibles en R y Python para calcular una matriz de confusión, pero en el caso binario, es sencillo calcular a mano (ídem).

```python
ConfMtz=metrics.confusion_matrix(y, modelLR2.predict(x))
type(ConfMtz)
```

```python
TN=ConfMtz[0,0]; TP=ConfMtz[1,1];FP=ConfMtz[0,1];FN=ConfMtz[1,0]
print('TN=',TN,' TP=',TP,' FP=',FP,' FN=',FN)
```

```python
print('Tasa de clasificación errónea:',(FP+FN)/(FP+FN+TP+TN))
```

El anterior resultado tiene sentido, ya que sumado con 'Accuracy' o tasa de clasificación acertada da como resultado casi 1.

```python
print("Tasa de Falsos Positivos '(Error tipo I)':",FP/(FP+TN))
```

```python
print("Tasa de Falsos Negativos '(Error tipo II)':",FN/(FN+TP))
```

```python
print("precisión (exactitud) ó tasa de predicciones negativas correctas:",TN/(TN+FN))
```

```python
print("precisión (exactitud) ó tasa de predicciones positivas correctas:",TP/(TP+FP))
```

```python
print("ESPECIFICIDAD ó 'recall' ó tasa de TN Negativos que se pedijeron como Negativos:",TN/(TN+FP))
```

```python
print("Tasa de verdaderos positivos o SENSIBILIDAD ó 'recall' ó (integridad):",TP/(FN+TP))
```

En estadística la capacidad de un sistema para dar con los valores verdaderos y evitar los valores falsos se llama sensibilidad (para los positivos) y especificidad (para los negativos) (wikipedia).

<A name='ROC'> </A> <h3>Curva ROC para visualizar el rendimiento de un clasificador binario</h3>

```python
print ("AUC :", metrics.roc_auc_score(y, modelLR2.predict_proba(x)[:,1]))
```

Una curva ROC (Receiver Operating Characteristic ó Característica Operativa del Receptor) es una métrica importante, una forma más utilizada para visualizar el rendimiento de un clasificador binario; y se cree que AUC (Area Under the Curve) es una de las mejores formas de resumir el rendimiento en un solo número. AUC indica que el clasificador calificará más alto la probabilidad de un ejemplo positivo seleccionado al azar que un ejemplo negativo seleccionado al azar. Si tiene varios modelos con casi la misma precisión, puede elegir el que proporcione un AUC más alto.

### Plot ROC curve

```python
# Determine the false positive and true positive rates
fpr, tpr, _ = metrics.roc_curve(y, modelLR2.predict_proba(x)[:,1])
```

```python
# Calculate the AUC
roc_auc = metrics.auc(fpr, tpr)
print ('ROC AUC: %0.2f' % roc_auc)
```

```python
# Plot of a ROC curve for a specific class
plt.figure()
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
```
